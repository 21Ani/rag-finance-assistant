```python
# convert words to numerical indices
def text_to_indices(text, vocab):

  indexed_text = []

  for token in tokenize(text):

    if token in vocab:
      indexed_text.append(vocab[token])
    else:
      indexed_text.append(vocab['<UNK>'])

  return indexed_text
```

```python
import torch
from torch.utils.data import Dataset
```

```python


class NewsDataset(Dataset):
    def __init__(self, df, vocab, max_length=50):
        """
        df        : pandas DataFrame with 'text' and 'label' columns
        vocab     : dict mapping tokens to indices
        max_length: max sequence length for padding/truncation
        """
        self.texts = df['text'].tolist()
        self.labels = df['label'].tolist()
        self.vocab = vocab
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        # Get text and label
        text = self.texts[idx]
        label = self.labels[idx]

        # Convert text to indices
        numerical_text = text_to_indices(text, self.vocab)

        # Pad or truncate
        numerical_text = self.pad_or_truncate(numerical_text)

        # Convert to tensors
        text_tensor = torch.tensor(numerical_text, dtype=torch.long)
        label_tensor = torch.tensor(label, dtype=torch.long)

        return text_tensor, label_tensor

    def pad_or_truncate(self, indices):
        """Pad with <PAD> token or truncate to max_length"""
        if len(indices) < self.max_length:
            indices += [self.vocab['<PAD>']] * (self.max_length - len(indices))
        else:
            indices = indices[:self.max_length]
        return indices
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.cuda.amp import GradScaler, autocast
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
class RNNClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_layers, num_classes, dropout):
        super(RNNClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(
            input_size=embedding_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            dropout=dropout if num_layers > 1 else 0.0,  # no dropout for single layer
            bidirectional=True,
            batch_first=True
        )
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim * 2, num_classes)  # *2 for bidirectional

    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, (h_n, c_n) = self.lstm(embedded)
        h_forward = h_n[-2, :, :]
        h_backward = h_n[-1, :, :]
        h_combined = torch.cat((h_forward, h_backward), dim=1)
        out = self.dropout(h_combined)
        out = self.fc(out)
        return out
```

```python
import itertools
from torch.cuda.amp import autocast, GradScaler

# Hyperparameter grid
param_grid = {
    'embedding_dim': [64, 128],
    'hidden_dim': [128, 256],
    'num_layers': [1, 2],
    'dropout': [0.3, 0.5],
    'learning_rate': [1e-3, 5e-4]
}
param_combinations = list(itertools.product(*param_grid.values()))
param_names = list(param_grid.keys())

best_acc = 0.0
best_params = None
vocab_size = 36146  # Replace with actual vocab size
num_classes = 20
```

```python

for params in param_combinations:
    # Unpack params
    config = dict(zip(param_names, params))
    print(f"\n🔄 Testing config: {config}")

    # Model, optimizer, loss
    model = RNNClassifier(
        vocab_size=vocab_size,
        embedding_dim=config['embedding_dim'],
        hidden_dim=config['hidden_dim'],
        num_layers=config['num_layers'],
        num_classes=num_classes,
        dropout=config['dropout']
    ).to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=config['learning_rate'], weight_decay=1e-2)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.9)
    scaler = GradScaler()

    # Train for few epochs (e.g., 5) just for tuning
    for epoch in range(5):
        model.train()
        total_loss = 0
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            with autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            total_loss += loss.item()
        scheduler.step()

    # Evaluate
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f"✅ Config Accuracy: {accuracy:.2f}%")

    # Track best
    if accuracy > best_acc:
        best_acc = accuracy
        best_params = config

print(f"\n🎯 Best Hyperparameters: {best_params} | Accuracy: {best_acc:.2f}%")
```

```python
# Build model with best hyperparameters
model = RNNClassifier(
    vocab_size=vocab_size,
    embedding_dim=best_params['embedding_dim'],
    hidden_dim=best_params['hidden_dim'],
    num_layers=best_params['num_layers'],
    num_classes=num_classes,
    dropout=best_params['dropout']
).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=best_params['learning_rate'], weight_decay=1e-2)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.9)
scaler = GradScaler()

# Train for 40 epochs
for epoch in range(40):
    model.train()
    total_loss = 0
    for inputs, labels in dataloader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        with autocast():
            outputs = model(inputs)
            loss = criterion(outputs, labels)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        total_loss += loss.item()
    avg_train_loss = total_loss / len(dataloader)

    # Evaluate
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    print(f"Epoch {epoch+1}/40 | Train Loss: {avg_train_loss:.4f} | Eval Acc: {accuracy:.2f}%")
    scheduler.step()
```

```python
import torch
import json

# 📦 Your label map
label_map = {
    0: "Analyst Update",
    1: "Fed | Central Banks",
    2: "Company | Product News",
    3: "Treasuries | Corporate Debt",
    4: "Dividend",
    5: "Earnings",
    6: "Energy | Oil",
    7: "Financials",
    8: "Currencies",
    9: "General News | Opinion",
    10: "Gold | Metals | Materials",
    11: "IPO",
    12: "Legal | Regulation",
    13: "M&A | Investments",
    14: "Macro",
    15: "Markets",
    16: "Politics",
    17: "Personnel Change",
    18: "Stock Commentary",
    19: "Stock Movement"
}

# 🔥 Tokenization helper
def tokenize(text, vocab, seq_len=50):
    tokens = [vocab.get(word.lower(), vocab["<UNK>"]) for word in text.split()]
    if len(tokens) < seq_len:
        tokens += [vocab["<PAD>"]] * (seq_len - len(tokens))  # Padding
    else:
        tokens = tokens[:seq_len]  # Truncate
    return torch.tensor([tokens], dtype=torch.long)  # Add batch dim

# 🎯 Prediction function
def predict_class(text, model, vocab, device, seq_len=50):
    model.eval()
    input_tensor = tokenize(text, vocab, seq_len).to(device)
    with torch.no_grad():
        outputs = model(input_tensor)  # shape: (1, num_classes)
        _, predicted_class = torch.max(outputs, 1)  # Get index of max logit
    class_idx = predicted_class.item()
    return label_map[class_idx]  # Map index to label
```

```python
user_input = input("📝 Enter English text for prediction:\n")
predicted_label = predict_class(user_input, model, vocab, device)
print(f"🔮 Predicted Category: {predicted_label}")
```
